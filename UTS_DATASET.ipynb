```python
import pandas as pd
```

```python
yo = pd.read_csv("/content/MQTT DoS Publish Flood.csv")
```

```python
yo
```

```python
ga = pd.read_csv("/content/MQTT Malformed.csv")
```

```python
ga
```

```python
yu = pd.read_csv("/content/Recon OS Scan.csv")
```

```python
yu
```

```python
wa = pd.read_csv("/content/Recon Ping Sweep.csv")
```

```python
wa
```

```python
hasilgabung = pd.concat([yo,ga,yu,wa], ignore_index=True)
```

```python
hasilgabung.columns.values
```

```python
Y = hasilgabung.iloc[:,7: 76]
```

```python
Y
```

```python
A = hasilgabung.iloc[:,83: 84]
```

```python
A
```

```python
from sklearn.model_selection import train_test_split
```

```python
Y_train,Y_test,A_train,A_test = train_test_split(Y,A, test_size = 0.2, random_state = 42)
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

```python
siti = DecisionTreeClassifier(criterion='entropy', splitter = 'random')
siti.fit(Y_train,A_train)
```

```python
A_pred = siti.predict(Y_test)
```

```python
A_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(A_test,A_pred)
```

```python
accuracy
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure(figsize = (10, 7))
tree.plot_tree(siti, feature_names = Y.columns.values, class_names = np.array([ 'MQTT Dos Publish Flood','MQTT Malformed','Recon OS Scan','Recon Pig Sweep']), filled = True)
plt.show()
```

```python
import seaborn as lol
from sklearn import metrics
label = np.array([ 'MQTT Dos Publish Flood','MQTT Malformed','Recon OS Scan','Recon Pig Sweep'])
```

```python
import matplotlib.pyplot as plt
conf_matrix = metrics.confusion_matrix(A_test, A_pred)
plt.figure(figsize=(10, 10))
lol.heatmap(conf_matrix, annot=True, xticklabels=label, yticklabels=label)
plt.xlabel('Prediksi')
plt.ylabel('Fakta')
plt.show()
```
